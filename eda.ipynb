```python
import pandas as pd
import matplotlib
import seaborn as sns
```

```python
df = pd.read_csv('data/data.csv', parse_dates=['DATE'], index_col='DATE', dayfirst=True)
```

```python
df
```

```python
df['PCT'] = df[['SPOT']].pct_change()
```

```python
sns.pairplot(df[['SPOT', 'TEMP']].dropna(), kind='reg')
```

```python
sns.heatmap(df[['SPOT', 'DXY', 'TEMP', 'EURUSD', 'SP500']].corr())
```
